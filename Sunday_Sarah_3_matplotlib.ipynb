## Task 1
Upload the data from https://bit.ly/2A2zkI6 into pandas DataFrame and display the first 10 rows.

```python
# your code here
import pandas as pd
#download data about films
df = pd.read_csv('https://raw.githubusercontent.com/Godoy/imdb-5000-movie-dataset/master/data/movie_metadata.csv'
            )
df.head(10)
```

## Task 2

## Task 3

```python
# your code here
# Extract frequency table for the content_rating column
frequency_table = df['content_rating'].value_counts()

# Display the frequency table
print(frequency_table)
```

## Task 4

Using previous data from previous exercise (use top-5 by frequency), create bar chart (sort columns in decreasing way)

```python
# your code here
import pandas as pd
import matplotlib.pyplot as plt

# Extract frequency table for the content_rating column
frequency_table = df['content_rating'].value_counts()

# Get the top 5 content ratings
top_5 = frequency_table.head(5)

# Sort the top 5 in decreasing order
top_5_sorted = top_5.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_5_sorted.plot(kind='bar', color='purple')
plt.title('Top 5 Content Ratings by Frequency')
plt.xlabel('Content Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y',linewidth= '0.5')

# Show the plot
plt.show()
```

## Task 5

Create the histogram of films' duration

```python
# your code here
import pandas as pd
import matplotlib.pyplot as plt

# Create a histogram of the duration column
plt.figure(figsize=(10, 6))
plt.hist(df['duration'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Films\' Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Films')

# Show the plot
plt.show()
```

## Task 6

Calculate the mean imdb_score for two film categories (color and black & white)

## Task 8

Calculate the number of unique genres

```python
# your code here
# Split the genres by the '|' delimiter and get the unique genres
unique_genres = df['genres'].str.split('|', expand=True).stack().unique()

# Calculate the number of unique genres
num_unique_genres = len(unique_genres)

# Display the result
print(f"The number of unique genres is: {num_unique_genres}")
```

## Task 9

Create scatterplot: x-axis-duration, y-axis-imdb_score, points'color-color

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean the 'color' column: strip spaces and handle NaN values
df['color'] = df['color'].str.strip().fillna('Black and White')

# Create a scatterplot with duration on x-axis, imdb_score on y-axis, and color differentiation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration', y='imdb_score', hue='color', data=df, palette={'Color': 'red', 'Black and White': 'blue'})
plt.title("Scatterplot of Film Duration vs IMDb Score")
plt.xlabel("Duration (minutes)")
plt.ylabel("IMDb Score")
plt.grid(True)
plt.show()
```

## Task 10

Calculate the frequency table of actor names (combine 3 columns for your analysis) for Action movie  genre. Display TOP-5 actors

```python
# Filter for Action movies 
action_movies = df[df['genres'].str.contains('Action', case=False, na=False)]

# Combine the actor columns into a single Series
all_actors = pd.concat([action_movies['actor_1_name'], action_movies['actor_2_name'], action_movies['actor_3_name']])

# Calculate the frequency of each actor
actor_frequency = all_actors.value_counts()

# Display the top 5 actors
top_5_actors = actor_frequency.head(5)
print(top_5_actors)
```
